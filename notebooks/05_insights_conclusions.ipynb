# Resumen de principales aprendizajes obtenidos a partir del análisis de datos realizado en  notebooks anteriores.

## Objetivo 

- Sintetizar resultados más importantes 
- Extracción y conclusiones prácticas sobre el comportamiento de usuarios.
- Uso de la plataforma y la generación de ingresos.


A partir del análisis exploratorio, se observa:

- La base de usuarios es diversa en cuanto a país, tipo de plan y canal de adquisición.
- Existen diferencias claras entre usuarios gratuitos y de pago.
- Algunos segmentos muestran mayor actividad y permanencia en la plataforma.

Esto indica que no todos los usuarios aportan el mismo valor al producto.


Del análisis de sesiones se pueden extraer las siguientes conclusiones:

- Existe una diferencia clara entre usuarios activos y usuarios ocasionales.
- Los usuarios más activos tienden a tener sesiones más frecuentes y largas.
- El comportamiento de uso varía según el plan contratado.

Esto sugiere que el nivel de engagement está relacionado con el tipo de usuario.


Ingresos:

- Una parte reducida de los usuarios genera la mayor parte del revenue.
- Algunas categorías de producto tienen un peso mayor en los ingresos totales.
- Los reembolsos existen, pero no representan la mayor parte del revenue.

Esto indica que la monetización está concentrada en determinados perfiles de usuario.


Análisis de churn:

- Una parte de los usuarios abandona la plataforma en fase temprana.
- Existen diferencias de retención entre los distintos planes.
- Los usuarios con mayor engagement tienden a permanecer más tiempo activos.

Esto refuerza la importancia de mejorar la experiencia en las primeras etapas del uso.


Resumen, y los principales insights del proyecto son:

- No todos los usuarios tienen el mismo valor para el negocio.
- El engagement está fuertemente relacionado con la retención.
- Determinados planes y segmentos generan mayor valor económico.
- Reducir el churn temprano puede tener un impacto significativo en los ingresos.


A partir del análisis realizado, se proponen las siguientes recomendaciones:

- Mejorar la experiencia de onboarding para reducir el churn temprano.
- Incentivar el uso recurrente de la plataforma mediante funcionalidades clave.
- Analizar en mayor profundidad los segmentos de mayor valor.
- Diseñar estrategias específicas para aumentar la retención de usuarios gratuitos.
